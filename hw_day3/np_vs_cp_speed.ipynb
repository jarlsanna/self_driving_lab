```python
import numpy as np
import cupy as cp
```

```python
n1 = 128
n2 = 256
n3 = 512
n4 = 1024
n5 = 2048
a = np.mgrid[:n1, :n1][0].astype(np.float32)
a2 = np.mgrid[:n2, :n2][0].astype(np.float32)
a3 = np.mgrid[:n3, :n3][0].astype(np.float32)
a4 = np.mgrid[:n4, :n4][0].astype(np.float32)
a5 = np.mgrid[:n5, :n5][0].astype(np.float32)

c = cp.mgrid[:n1, :n1][0].astype(cp.float32)
c2 = cp.mgrid[:n2, :n2][0].astype(cp.float32)
c3 = cp.mgrid[:n3, :n3][0].astype(cp.float32)
c4 = cp.mgrid[:n4, :n4][0].astype(cp.float32)
c5 = cp.mgrid[:n5, :n5][0].astype(cp.float32)
```

```python
%timeit np.fft.fft2(a)
```

```python
%timeit np.fft.fft2(a2)
```

```python
%timeit np.fft.fft2(a3)
```

```python
%timeit np.fft.fft2(a4)
```

```python
%timeit np.fft.fft2(a5)
```

```python
%timeit cp.fft.fft2(c)
```

```python
%timeit cp.fft.fft2(c2)
```

```python
%timeit cp.fft.fft2(c3)
```

```python
%timeit cp.fft.fft2(c4)
```

```python
%timeit cp.fft.fft2(c5)
```

For all sizes cupy was faster. With 128x128, cupy was almost twice as fast and for size 2048x2048 was around 1000 times faster ~100 miliseconds vs 100 microseconds...

Changing the data type from int64 to float 32, didnt have a significant impact on numpy fft2, but made cupy operations much slower, for 2048x2048 cupy now took 1 milisecond to compute.
